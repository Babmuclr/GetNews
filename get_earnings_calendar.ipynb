# 決算情報をinvesting.comからとる方法
- seleniumとbeautifulSoupを用いて行う

```python
import urllib
from urllib import request  # urllib.requestモジュールをインポート
from bs4 import BeautifulSoup  # BeautifulSoupクラスをインポート
```

```python
import datetime as dt 
now = dt.datetime.utcnow()
Month = now.strftime("%B")
```

```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
```

```python
# 目的のURL
URL = 'https://www.investing.com/earnings-calendar/' 
```

```python
#　ヘッドレスモードでブラウザを起動
options = Options()
# optionsによって、seleniumでGoogleChromeを表示したくない場合
# options.add_argument('--headless') 
    
# ブラウザーを起動
# ドライバーの選択
# MACOS
# driver = webdriver.Chrome('./chromedriver', options=options)
# M1

driver = webdriver.Chrome('./chromedriver_m1', options=options)
driver.get(URL)
driver.implicitly_wait(10)
```

# ここで一旦ストップ
決算情報のほしい期間だけを、手動で指定する

```python
html = driver.page_source
```

```python
soup = BeautifulSoup(html, "html.parser")
soup = soup.find(id="earningsCalendarData")
```

```python
data = soup.find_all("td")
x = []
for i in data:
    x.append(i.getText())
```

```python
def change_text(x):
    if x =="\n\n" or x == "" or x == "\xa0":
        return "NoData"
    elif "\xa0" in x:    
        x = x.replace("\xa0", "")
        if "\n" in x:
            x = x.replace("\n", "")
            if "                    " in x:
                x = x.replace("                    ", "")
        elif "/" in x:
            x = x.replace("/", "")
    return x
```

```python
data = list(map(change_text,x))
li = [i for i in data if i != "NoData"]
li
```

```python
driver.quit()
```

```python
li2 = []
day = ""
for i in li:
    if Month in i:
        day = i
    else:
        li2.append([day, i])
li2
```

```python
name_li = ["Company","EPS(Reported)","EPS(Estimated)","Revenue(Reported)","Revenue(Estimated)","MarketCap"]
ans = []
for num in range(len(li2) // 6):
    dic = {}
    dic["Time"] = li2[6 * num][0]
    for i in range(6):
        dic[name_li[i]] = li2[6 * num + i][1]
    ans.append(dic)
ans
```

```python
with open("./stocks.txt", mode="w") as f:
    for i in ans:
        f.write(i["Company"]+ "\n")
```
